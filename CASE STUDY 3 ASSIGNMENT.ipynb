```python
# importing libraries

import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize = (8, 5))
```

## 1. Line Plot

```python
# assigning the varibles

x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 15, 7]
```

```python
# creating the line plot

plt.plot(x, y)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line Plot")

plt.show()
```

## 2. Barplots

```python
# assigning the variables

categories = ['A', 'B', 'C', 'D']
values = [25, 40, 15, 30]
```

```python
# creating the barplot
plt.bar(categories, values, color='olive', width=0.4)

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Barplot")

plt.show()
```

## 3. Histogram

```python
# assigning the variables

data = [32, 45, 50, 28, 36, 42, 38, 25, 29, 52, 48, 60]
```

```python
# creating the histogram

plt.hist(data, color='lime', edgecolor='black')

plt.xlabel("Data")
plt.ylabel("Frequency")
plt.title("Histogram")

plt.show()
```

## 4. Scatterplot

```python
# assigning the variables

x = [3, 5, 7, 10, 12, 15]
y = [25, 40, 30, 50, 35, 60]
```

```python
# creating the scatterplot

plt.scatter(x, y)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatterplot")

plt.show()
```

## 5. Pie Charts

```python
# assigning the variables

labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sizes = [30, 25, 20, 15]
```

```python
# creating the pie chart

plt.pie(sizes, labels=labels)

plt.show()
```

## 6. Boxplot

```python
# assigning the variables

data = [45, 60, 75, 300, 80, 350, 90, 95, 100, 110, 120, 250, 200]
```

```python
# creating the boxplot

plt.boxplot(data)

plt.show()
```

## 7. Barhplot

```python
# assigning the variables

categories = ['A', 'B', 'C', 'D']
values = [25, 40, 15, 30]
```

```python
# creating the barhplot

plt.barh(categories, values, color='pink')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Barh Plot')

plt.show()
```
